# 11. Introdução ao `iloc`

O `iloc` é usado para selecionar dados pela **posição numérica**.

Estrutura geral:

```python
df.iloc[linhas, colunas]
```

Aqui usamos posições, não nomes.

## Selecionando as 5 primeiras linhas e as 3 primeiras colunas

````python
df.iloc[0:5, 0:3]
````

Esse código significa:

> Pegue as linhas da posição 0 até antes da posição 5 e as colunas da posição 0 até antes da posição 3.

## Diferença entre `loc` e `iloc`

| Recurso | Seleciona por | Exemplo |
|---|---|---|
| `loc` | nome da linha/coluna | `df.loc[:, ["gender", "math_score"]]` |
| `iloc` | posição numérica | `df.iloc[0:5, 0:3]` |

Nesta aula, vamos usar `loc` e `iloc` apenas de forma introdutória. O aprofundamento pode aparecer mais adiante, em limpeza de dados e EDA.

# 16. Erro comum: esquecer os parênteses

No Pandas, quando usamos `&` ou `|`, cada condição precisa estar entre parênteses.

Forma correta:

```python
df[(condicao1) & (condicao2)]
```

Forma incorreta:

```python
df[condicao1 & condicao2]
```

# 27. Perguntas analíticas com filtros

Agora vamos responder algumas perguntas usando os comandos aprendidos.

## Pergunta 1

Quais estudantes tiveram média geral maior ou igual a 80?

````python
#Codigo aqui
df.columns = df.columns.str.replace(" ", "_").str.replace("/", "_")

df["media_geral"] = (df["math_score"] + df["writing_score"] + df["reading_score"]) / 3
print(df[(df["media_geral"] >= 80)])
````

## Pergunta 2

Quais estudantes fizeram curso preparatório e tiveram média geral maior ou igual a 80?

````python
# Codigo aqui
print(df[(df["media_geral"] >= 80) & (df["test_preparation_course"] == "completed")])

````

## Pergunta 3

Quais estudantes não fizeram curso preparatório e foram reprovados?

````python
# Codigo aqui
def situacao_estudante(media):
     if media >= 60:
         return "Aprovado"
     else:
         return "Reprovado"

df["situacao"] = df["media_geral"].apply(situacao_estudante)

print(df[(df["test_preparation_course"] == "none") & (df["situacao"] == "Reprovado")])
````

## Pergunta 4

Quais estudantes tiveram nota de matemática menor que 50, mas nota de leitura maior ou igual a 70?

````python
#codigo aqui
print(df[(df["math_score"] < 50) & (df["reading_score"] >= 70)])
````
